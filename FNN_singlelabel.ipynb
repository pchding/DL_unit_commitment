```python
import torch
import torch.nn as nn
import torch.utils.data
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
i1 = pd.read_csv("inputf2.csv", index_col = 0)
i2o = pd.read_csv('tlo2.csv', index_col = 0)
```

```python
i2 = i2o.loc[:,'l118']
```

```python
i2[i2 == 2] = 0
```

```python
X_train, X_test, y_train, y_test = train_test_split(i1, i2, train_size = 0.8, test_size = 0.2)
Xt_train = torch.from_numpy(X_train.values).double()
Xt_test = torch.from_numpy(X_test.values).double()
yt_train = torch.from_numpy(y_train.values).double()
yt_test = torch.from_numpy(y_test.values).double()

train_dataset = torch.utils.data.TensorDataset(Xt_train,yt_train)
test_dataset = torch.utils.data.TensorDataset(Xt_test,yt_test)
```

```python
train_loader = torch.utils.data.DataLoader(dataset=train_dataset, 
                                           shuffle=True)

test_loader = torch.utils.data.DataLoader(dataset=test_dataset,
                                          shuffle=False)
```

```python
class simple1(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(simple1, self).__init__()
        # Linear function 1: 784 --> 100
        self.fc1 = nn.Linear(input_dim, hidden_dim) 
        # Non-linearity 1
        self.relu1 = nn.ReLU()

        # Linear function 2: 100 --> 100
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        # Non-linearity 2
        self.relu2 = nn.ReLU()

        # Linear function 3: 100 --> 100
        self.fc3 = nn.Linear(hidden_dim, hidden_dim)
        # Non-linearity 3
        self.relu3 = nn.ReLU()

        # Linear function 4 (readout): 100 --> 10
        self.fc4 = nn.Linear(hidden_dim, output_dim)  
        

    def forward(self, x):
        # Linear function 1
        out = self.fc1(x)
        # Non-linearity 1
        out = self.relu1(out)

        # Linear function 2
        out = self.fc2(out)
        # Non-linearity 2
        out = self.relu2(out)

        # Linear function 2
        out = self.fc3(out)
        # Non-linearity 2
        out = self.relu3(out)

        # Linear function 4 (readout)
        out = self.fc4(out)
        
    
        return out
```

```python

input_dim = 96
hidden_dim = 50
output_dim = 2

model = simple1(input_dim, hidden_dim, output_dim)


criterion = nn.CrossEntropyLoss()



```

```python
num_epochs = 100
learning_rate = 0.00002
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

iter = 0
for epoch in range(num_epochs):
    losses = []
    for i, (data, labels) in enumerate(train_loader):
        # Load images with gradient accumulation capabilities
        data = data.view(-1, 96).requires_grad_().float()

        # Clear gradients w.r.t. parameters
        optimizer.zero_grad()

        # Forward pass to get output/logits
        outputs = model(data)

        # Calculate Loss: softmax --> cross entropy loss
        loss = criterion(outputs, labels.long())

        # Getting gradients w.r.t. parameters
        loss.backward()

        # Updating parameters
        optimizer.step()
        
        losses.append(loss.data.mean())
        print('[%d/%d] Loss: %.3f' % (epoch+1, num_epochs, np.mean(losses)))
```

```python
def getcat(output):
    top_n, top_i = output.topk(1)
    cat = top_i[0].item()
    return cat
```

```python
predy = np.full((yt_test.size()[0],1), 4)
testy = np.full((yt_test.size()[0],1), 4)
for i, (data, labels) in enumerate(test_dataset):
    data = data.view(-1, 96).float()
    out1 =  model(data)
    out1 = getcat(out1)
    predy[i] = out1
    testy[i] = labels.numpy()
    
```

```python
correct = 0
for i in range(predy.shape[0]):
    if np.array_equal(predy[i],testy[i]):
        print(i)
        correct += 1
testaccu = correct/predy.shape[0]
print(testaccu)
```

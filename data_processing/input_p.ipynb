```python
import pandas as pd
import numpy as np
```

```python
tlmin =  pd.read_csv("tlamin.csv", header=None, names = ['day','hour','line','status'])
tlmax =  pd.read_csv("tlamax.csv", header=None, names = ['day','hour','line','status'])
```

```python
tlmin.loc[:,'status'] = -2
```

```python
tlt = tlmin.loc[tlmin.day == 'd1']
s1 = tlt.loc[tlt.line== 'l28'].status.iloc[0]
s0 = 0
```

```python
llist = 'l' + pd.Series(range(1,121)).astype(str)
dlist = 'd' + pd.Series(range(1,367)).astype(str)
```

```python
olist = pd.DataFrame(columns = ['day'])
olist = pd.concat([olist,pd.DataFrame(columns = llist)],sort=False)
olist.day = dlist
olist = olist.fillna(int(0))
```

```python
olist.index = dlist
```

```python
for d in dlist:
    tlmint = tlmin.loc[tlmin.day == d]
    tlmaxt = tlmax.loc[tlmax.day == d]
    for l in llist:
        try:
            s1 = tlmint.loc[tlmint.line== l ].status.iloc[0]
        except IndexError:
            s1 = 0
        try:
            s2 = tlmaxt.loc[tlmaxt.line== l ].status.iloc[0]
        except IndexError:
            s2 = 0
        olist.loc[d,l] = olist.loc[d,l] + s1 + s2
    
```

```python
olist
```

```python
olist.to_csv("tlo.csv",index = False,header = True,na_rep="")
```

```python
# remove not interesting labels
(olist.l2 == 0).sum() == 366
for l in llist:
    if (olist.loc[:,l] == 0).sum() == 366:
        olist = olist.drop(columns = l)
```

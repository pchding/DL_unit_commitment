```python
import pandas as pd
import numpy as np
```

```python
#load_o = pd.read_csv("REAL_TIME_regional_load.csv")

load_o = pd.read_csv("RTS_wind_raw.csv")
```

```python
#define the lengths of time periods in minutes
tl = 30
nt = 24*60/tl

n0 = tl /(24*60/max(load_o.Period))
nr = load_o.shape[0]/n0
```

```python
load_n = pd.DataFrame(columns = load_o.columns)
```

```python
for x in range(int(nr)):
    m = x+1
    aux = load_o.iloc[int((m-1)*n0):int(m*n0),:].mean()
    load_n = load_n.append(aux, ignore_index=True)
```

```python
v = pd.Series(range(1,int(nt)+1))
load_n['Period'] = np.tile(v, len(load_n) // len(v) + 1)[:len(load_n)]  
```

```python
load_n['Year'] = load_n['Year'].astype('int')
load_n['Month'] = load_n['Month'].astype('int')
load_n['Day'] = load_n['Day'].astype('int')
```

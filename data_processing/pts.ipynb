```python
import pandas as pd
from fbprophet import Prophet
import datetime 
  

```

```python
data_dfo = pd.read_csv("RTS_wind_raw.csv")
data_df = data_dfo.copy()
data_df["Hour"] = 5*(data_df["Period"]-1)
data_df["Hour"]=data_df["Hour"].apply(lambda x: str(datetime.timedelta(minutes = x)))
data_df["YMD"] = pd.to_datetime(data_df.Year*10000+data_df.Month*100+data_df.Day,format='%Y%m%d').astype(str)

```

```python
import random
import pandas as pd
import numpy as np
```

```python
load_o = pd.read_csv("demand_30_raw.csv")
#load_o = pd.read_csv("DAY_AHEAD_regional_Load.csv")
load_dis = pd.read_csv("load_d.csv")
```

```python
load_o['Nday'] = load_o['Day']
```

```python
months = load_o.Month.unique()
months = months[1:]
db = 0
for m in months:
    db += max(load_o[load_o.Month == (m-1)].Day)
    load_o.loc[load_o.Month == m,'Nday'] = load_o[load_o.Month == (m)].Day + db
```

```python
load_o.Nday =  "d" + load_o.Nday.astype(str)
```

```python
load_n = pd.DataFrame(index = load_o.index)
```

```python
load_n['Day'] = load_o.Nday
#load_n["dot1"] = len(load_n)*["."]
load_n['Time'] = "t" + load_o.Period.astype(str)
load_n = pd.concat([load_n,pd.DataFrame(columns = load_dis.columns)],sort=False)
```

```python
np.random.seed(100)
rmat = pd.DataFrame(data = ((np.random.rand(load_n.shape[0],load_dis.shape[1])*2-1)*0.005),index = load_o.index, columns = load_dis.columns)
```

```python
load_n.loc[:,'s101':'s120'] = (rmat.loc[:,'s101':'s120']+1).mul(load_o.loc[:,'1'], axis =0) 
```

```python
load_n.loc[:,'s201':'s220'] = (rmat.loc[:,'s201':'s220']+1).mul(load_o.loc[:,'2'], axis =0) 
```

```python
load_n.loc[:,'s301':'s320'] = (rmat.loc[:,'s301':'s320']+1).mul(load_o.loc[:,'3'], axis =0) 
```

```python
load_n.loc[:,'s101':'s320'] = load_n.loc[:,'s101':'s320'].mul(load_dis.loc[0,:], axis =1)
```
